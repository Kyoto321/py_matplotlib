```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### Basic Graph

```python
## sometimes always refer to the documentation #matplotlib.org
```

```python

([0,1,2,3,4],[0,2,4,6,8]) #arrays

# we can move how arrays into variables

x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize the graph -- to specify the pixels per inch - pass in param='dpx'
plt.figure(figsize=(5,3), dpi=200)

# Line 1
# customization
# Use shorthand notation  #see marker documentation
# fmt = '[color][marker][line]'
plt.plot(x,y, 'b^--', label='2x')

## or

# Keyword argument notation
#plt.plot(x,y, label='2x', color='grey', linewidth=2, marker='.', linestyle='--', markersize=7, markeredgecolor='blue')

# Line 2
# Select intervals we want to plot points againts
x2= np.arange(0,4.5,0.5)

# Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'r', label='x^2')

#Plot reminder of graph as a dot
plt.plot(x2[5:], x2[5:]**2, 'r--')


# To add title to the graph / font,
plt.title('Our first graph', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 10})

# Adding a label
plt.xlabel('X Axis')
plt.ylabel('Y Axis') 

# to adjust decimal values on axis/ and set the dimensions
plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])

3# Adding a Legend
plt.legend()
 
# To save the graph dpi=300 good for saving graphs high resolutions
plt.savefig('mygraph.png', dpi=300)

# show plot 
plt.show()
```

### Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values)

bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch('*')

plt.figure(figsize=(6,4))

plt.show()
```
